## Converters:

```python
def converter_for_day(col):
    if col=='NaN':
        return 0
    else:
        return col
def converter_for_wind_speed(col):
    if col=='Not available':
        return 0
    else:
        return col


df=pd.read_csv("D:/Courses/Data-analysis-with-pandas-master/datasets/season.csv",converters={
    'day':converter_for_day,
    'wind-speed':converter_for_wind_speed
   })
df
```
